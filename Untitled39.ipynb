```python
import cv2
import torch
import torchvision.transforms as transforms
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from google.colab.patches import cv2_imshow

# Load the pre-trained Faster R-CNN model
model = fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()

# Set up transformation for input images
transform = transforms.Compose([
    transforms.ToTensor(),
])

# Open video capture
video_path = "/content/faces01_j2M9XC65.mp4"
cap = cv2.VideoCapture(video_path)

# Get video dimensions
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Initialize counters
total_faces = 0
face_counter = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to RGB (as the model expects)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Apply transformations
    input_tensor = transform(rgb_frame).unsqueeze(0)

    # Make prediction
    with torch.no_grad():
        prediction = model(input_tensor)

    # Get bounding boxes and labels
    boxes = prediction[0]['boxes'].cpu().numpy()
    labels = prediction[0]['labels'].cpu().numpy()

    # Draw bounding boxes and save detected faces as images
    for box, label in zip(boxes, labels):
        if label == 1:  # Assuming label 1 corresponds to the "person" class
            total_faces += 1
            box = [int(i) for i in box]
            cv2.rectangle(frame, (box[0], box[1]), (box[2], box[3]), (0, 255, 0), 2)
            if face.shape[0] > 0 and face.shape[1] > 0:
                face_counter += 1
                face_filename = f"face_{face_counter}.jpg"
                cv2.imwrite(face_filename, face)
                print(f"Face captured: {face_filename}")

    # Add overlay text
    overlay_text = f"Face detected: {face_counter}, Total Faces: {total_faces}"
    cv2.putText(frame, overlay_text, (frame_width - 400, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

    # Display the frame
    cv2_imshow(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
